## Second Attempt to Interpolate the data

The goal is to **Time interpolate** each region/loction in a sepearet data frame; according to the Class and Sorting by Questions to avoid wrong interpolations due to missing years in specific questions.

Steps:
1. Create a df for each region/location. (5 dfs in total)
2. Set YearStart as the index.
3. Sort by Questions column.
4. Group by Class,YearStart
5. Reset Indexes for each df.
6. Combine all dataframes in a one dataset (5 dfs in total)

```python
import pandas as pd
```

```python
org_df = pd.read_excel('Clean_alzh.xlsx',sheet_name='Clean_alzh') #read the original df
```

```python
 org_df.drop(columns='Column1',inplace=True) #dropping the index column
```

```python
location = list(org_df['Geographical Area'].unique()) #get the list of the locations
```

```python
location #sanity check
```

```python
#Create a seperate Df for each location

Midwest_df = pd.DataFrame()
Northeast_df = pd.DataFrame()
South_df = pd.DataFrame()
US_df = pd.DataFrame()
West_df = pd.DataFrame()
```

```python
for loc in location:
    if loc == 'Midwest':
        Midwest_df = org_df[org_df['Geographical Area'] == loc]
    elif loc == 'Northeast':
        Northeast_df = org_df[org_df['Geographical Area'] == loc]
    elif loc == 'South':
        South_df = org_df[org_df['Geographical Area'] == loc]
    elif loc == 'West':
        West_df = org_df[org_df['Geographical Area'] == loc]
    elif loc == 'United States, DC & Territories':
        US_df = org_df[org_df['Geographical Area'] == loc]
    else:
        print('Location Not Found!')
```

```python
#changing the YearStart value for a datetime before time interpolation

Midwest_df['YearStart'] = pd.to_datetime(Midwest_df['YearStart'], format='%Y')
Northeast_df['YearStart'] = pd.to_datetime(Northeast_df['YearStart'], format='%Y')
South_df['YearStart'] = pd.to_datetime(South_df['YearStart'], format='%Y')
West_df['YearStart'] = pd.to_datetime(West_df['YearStart'], format='%Y')
US_df['YearStart'] = pd.to_datetime(US_df['YearStart'], format='%Y')
```

```python
#time interpolating the Data Value column for each Class

Midwest_intr = pd.DataFrame(Midwest_df.groupby('Question')['Data_Value'].apply(lambda x: x.interpolate(method='time')))
Northeast_intr = pd.DataFrame(Northeast_df.groupby('Question')['Data_Value'].apply(lambda x: x.interpolate(method='time'))) 
West_intr = pd.DataFrame(West_df.groupby('Question')['Data_Value'].apply(lambda x: x.interpolate(method='time')))
South_intr = pd.DataFrame(South_df.groupby('Question')['Data_Value'].apply(lambda x: x.interpolate(method='time')))
US_intr = pd.DataFrame(US_df.groupby('Question')['Data_Value'].apply(lambda x: x.interpolate(method='time')))
```

```python
#resetting the index for each df to access the interpolated Data Value column in each region df

Midwest_intr.reset_index(inplace=True)
Northeast_intr.reset_index(inplace=True)
West_intr.reset_index(inplace=True)
South_intr.reset_index(inplace=True)
US_intr.reset_index(inplace=True)
```
