## Adding the Population to each location

```python
#testing iterating rows in the alzheimers dataset
for index, row in alzh_df2.iterrows():
    print(index,row['Geographical Area'], row['YearStart'])
    if index == 200:
        break
    
```

```python
#function to get the population for states in each year
def get_pop(state,year):
    pop = 0
    for index, row in regions_df.iterrows():
        if row['Geographical Area'] == state:
            if year == 2015:
                pop = row['2015']
                break
            if year == 2016:
                pop = row['2016']
                break
            if year == 2017:
                pop = row['2017']
                break
            if year == 2018:
                pop = row['2018']
                break
            if year == 2019:
                pop = row['2019']
                break
            if year == 2020:
                pop = row['2020']
                break
            if year == 2021:
                pop = row['2021']
                break
            if year == 2022:
                pop = row['2022']
                break
    return pop
```

```python
get_pop('All regions',2020) #sanity check
```

```python
#a loop to iterate through the rows and get the population for each row, saved in a list
ls = []

for index, row in alzh_df2.iterrows():
    ls.append(get_pop(row['Geographical Area'],row['YearStart']))
```
